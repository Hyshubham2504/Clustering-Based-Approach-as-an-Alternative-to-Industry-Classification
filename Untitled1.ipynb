### 3. Get the daily return data of the 500 companies from Yahoo Finance
For the 20-year period 2004-01-01 — 2024-01-01 (YYYY-MM-dd).

```python
# Temporary fix: Yahoo Finance uses "BRK-B" instead of "BRK.B" and "BF-B" instead of "BF.B"
# Replace these 2 tickers in the SP500_tickers list so that we can get their data from Yahoo Finance
SP500_tickers = list(map(lambda x: x.replace('BRK.B', 'BRK-B'), SP500_tickers))
SP500_tickers = list(map(lambda x: x.replace('BF.B', 'BF-B'), SP500_tickers))
```

```python
# This can take 3-4 minutes
start = datetime.now()
SP500_dict = {}

for ticker in SP500_tickers:
    ticker_yahoo = yf.Ticker(ticker)
    adj_close_data = ticker_yahoo.history(start="2004-01-01", end="2024-01-01", auto_adjust=False, actions=False)["Adj Close"]
    if not adj_close_data.empty:
        SP500_dict[ticker] = adj_close_data.pct_change().dropna()

print(datetime.now() - start)
```

```python
daily_return_df = pd.DataFrame(SP500_dict)
descriptive_data = daily_return_df.describe()
descriptive_data
```

### Checking for rolling window periods(return mean and std)

1. Weekly Rolling Window: Approximately 5 trading days.
2. Monthly Rolling Window: Approximately 21 trading days.
3. Quarterly Rolling Window: Approximately 63 trading days.
4. Half-Yearly Rolling Window: Approximately 126 trading days.
5. Yearly Rolling Window: Approximately 252 trading days.

```python
# Define rolling windows in terms of trading days
rolling_windows = {
    'Weekly': 5,
    'Monthly': 21,
    'Quarterly': 63,
    'Half-Yearly': 126,
    'Yearly': 252
}

# Select a diverse set of companies from different sectors
diverse_companies = daily_return_cleaned_df.columns[:10] 

rolling_stats = {}
for period, window in rolling_windows.items():
    rolling_mean = daily_return_cleaned_df[diverse_companies].rolling(window=window).mean()
    rolling_std = daily_return_cleaned_df[diverse_companies].rolling(window=window).std()
    rolling_stats[period] = {
        'Mean': rolling_mean,
        'Std': rolling_std
    }

plt.figure(figsize=(20, 25))
for i, company in enumerate(diverse_companies, 1):
    plt.subplot(5, 2, i)
    for period in rolling_windows:
        plt.plot(rolling_stats[period]['Mean'][company], label=f'{period} Mean')
    plt.title(f'Rolling Mean for {company}')
    plt.xlabel('Date')
    plt.ylabel('Rolling Mean')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()

plt.figure(figsize=(20, 25))

for i, company in enumerate(diverse_companies, 1):
    plt.subplot(5, 2, i)
    for period in rolling_windows:
        plt.plot(rolling_stats[period]['Std'][company], label=f'{period} Std')
    plt.title(f'Rolling Std for {company}')
    plt.xlabel('Date')
    plt.ylabel('Rolling Std')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
## The green line(quarterly std and return mean looks more stable, balanced and less noise) - 63 days
## And we will also consider monthly returns to capture short term trends or fluctuations- 21 days
```

### Using beta 
Beta measures a stock's volatility relative to the overall market. A beta greater than 1 indicates that the stock is more volatile than the market, while a beta less than 1 indicates lower volatility.

#### std deviation also meaasure volatility but then why beta????
1. Beta: Measures the stock's sensitivity to overall market movements. It indicates how much the stock's price is expected to change in response to changes in the market.
Standard Deviation: Measures the stock's overall volatility, regardless of market movements. It indicates how much the stock's price varies over time.
2. Combining Beta with other features (like rolling mean and standard deviation) will help to differentiate between stocks based on both market-related and stock-specific factors.

```python
market_data = daily_return_df.mean(axis=1)
```

```python
market_data
```

```python
daily_return_df_clean = daily_return_df.dropna(axis=1)

# Calculate Beta for each stock
beta = daily_return_df_clean.apply(lambda x: np.cov(x, market_data)[0, 1] / np.var(market_data))    
```
